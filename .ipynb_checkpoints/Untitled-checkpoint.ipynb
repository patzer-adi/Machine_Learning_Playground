```python
import torch
import matplotlib.pyplot as plt

# Reproducibility
torch.manual_seed(0)

# Parameters
num_samples = 100
true_w, true_b = 2.0, -1.0
noise_std = 0.7

# Features
X = torch.rand(num_samples)

# Noise
noise = torch.randn(num_samples) * noise_std

# Labels
y = true_w * X + true_b + noise

# Plot
plt.scatter(X.numpy(), y.numpy(), label="Synthetic data", alpha=0.7)
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
w = torch.tensor(0.0)
b = torch.tensor(0.0)

epochs = 50

lr = 0.05

losses = []

for epoch in range(epochs):
    for i in range(num_samples):
        xi = X[i]
        yi = y[i]

        y_pred = w*xi + b
        loss = (y_pred - y) ** 2

        losses.append(loss)

        grad_w = 2*(y_pred - yi) * xi
        grad_b = 2*(y_pred - yi)

        w = w - lr * grad_w
        b = b - lr * grad_b

plt.scatter(X.numpy(), y.numpy(), label="Synthetic data", alpha=0.7)
plt.plot(X.numpy(),(w*X = b))
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data")
plt.legend()
plt.show()
```
