```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
# we set target variables as 0 = maligant, 1 is benin
```

```python
features = ['mean radius', 'mean texture', 'mean_smoothness']
#uni variate analysis that is it takes one feature at a time and explores them
```

```python
df[features].hist()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
df[features].boxplot()
plt.show()
```

```python
#it shows that there are outliers in the data
```

```python
import seaborn as sns

for feature in features:
    sns.boxplot(x='target', y=feature, data=df)
    plt.title(feature)
    plt.show()
```

```python
sns.scatterplot(x='mean radius', y = 'mean texture', data = df)
plt.show()
```

```python
#correaltion analysis

corr = df.drop('target', axis = 1).corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap = 'coolwarm')
plt.show()
```
